```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('life_expectancy.csv')
```

```python
data.head()
```

```python
kazakhstan_data = data[data['Entity'] == 'Kazakhstan']
```

```python
kazakhstan_data = kazakhstan_data[:-7]
```

```python
df = pd.DataFrame(kazakhstan_data)
```

```python
kazakhstan_data
```

```python
df.to_excel('kazakhstan_data.xlsx', index=False)
```

```python
plt.plot(kazakhstan_data['Year'], kazakhstan_data['LifeExpectancy'])
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy in Kazakhstan from 1950 to 2021')
plt.grid(True)
plt.savefig('life_expactancy_in_kazakhstan.png')
plt.show
```

```python
data_ages = pd.read_csv('life_expectancy_different_ages.csv')
```

```python
data_ages
```

```python
kazakhstan_ages = data_ages[data_ages['Entity'] == 'Kazakhstan']
```

```python
kazakhstan_ages = kazakhstan_ages[:-7]
```

```python
data_ages[['LifeExpectancy0', 'LifeExpectancy10', 'LifeExpectancy25', 'LifeExpectancy45', 'LifeExpectancy65', 'LifeExpectancy80']] = data_ages[['LifeExpectancy0', 'LifeExpectancy10', 'LifeExpectancy25', 'LifeExpectancy45', 'LifeExpectancy65', 'LifeExpectancy80']].apply(lambda x: x.astype(float).round(2))
```

```python
kazakhstan_ages
```

```python
kazakhstan_ages.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy Trends in Kazakhstan')
plt.savefig('life_expectancy_different_ages_in_kazakhstan')
plt.show()
```
